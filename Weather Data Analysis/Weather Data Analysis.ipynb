```python
import pandas as pd
df = pd.read_csv('Weather Data.csv')

# Initial exploration
print("--- Dataset Overview ---")
print(df.head())
print(f"\nShape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")


# Q. 1) Unique 'Wind Speed' values
unique_wind_speeds = df['Wind Speed_km/h'].unique()

# Q. 2) Number of times 'Weather' is exactly 'Clear'
weather_clear_count = df[df['Weather'] == 'Clear'].shape[0]

# Q. 3) Number of times 'Wind Speed was exactly 4 km/h'
wind_speed_4_count = df[df['Wind Speed_km/h'] == 4].shape[0]

# Q. 4) Null Values in the data
null_values = df.isnull().sum()

# Q. 5) Rename 'Weather' to 'Weather Condition'
# Using inplace=True to modify the dataframe permanently
df.rename(columns={'Weather': 'Weather Condition'}, inplace=True)

# Q. 6) Mean 'Visibility'
mean_visibility = df['Visibility_km'].mean()

# Q. 7) Standard Deviation of 'Pressure'
std_pressure = df['Press_kPa'].std()

# Q. 8) Variance of 'Relative Humidity'
var_humidity = df['Rel Hum_%'].var()

# Q. 9) All instances when 'Snow' was recorded
snow_records = df[df['Weather Condition'].str.contains('Snow')]

# Q. 10) Instances when 'Wind Speed is above 24' and 'Visibility is 25'
wind_vis_filter = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)]

# Q. 11) Mean value of each column against each 'Weather Condition'
mean_per_condition = df.groupby('Weather Condition').mean(numeric_only=True)

# Q. 12) Min & Max value of each column against each 'Weather Condition'
min_max_per_condition = df.groupby('Weather Condition').agg(['min', 'max'])

# Q. 13) All Records where Weather Condition is Fog
fog_records = df[df['Weather Condition'] == 'Fog']

# Q. 14) Instances when 'Weather is Clear' or 'Visibility is above 40'
clear_or_vis_40 = df[(df['Weather Condition'] == 'Clear') | (df['Visibility_km'] > 40)]

# Q. 15) Complex Filtering:
# A. 'Weather is Clear' and 'Rel Hum > 50' OR B. 'Visibility is above 40'
complex_filter = df[((df['Weather Condition'] == 'Clear') & (df['Rel Hum_%'] > 50)) | (df['Visibility_km'] > 40)]


print("\n--- Specific Analysis Results ---")
print(f"1. Unique Wind Speeds: {unique_wind_speeds[:5]}... (total {len(unique_wind_speeds)})")
print(f"2. Times Weather was Clear: {weather_clear_count}")
print(f"6. Mean Visibility: {mean_visibility:.2f} km")
print(f"15. Complex Filter Result Rows: {len(complex_filter)}")

# To see the Snow records specifically:
# print(snow_records.head())
```
