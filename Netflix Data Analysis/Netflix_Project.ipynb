```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv('Netflix Dataset.csv')

# Task 1: Remove Duplicates
df.drop_duplicates(inplace=True)

# Task 2: Null Values Heatmap
sns.heatmap(df.isnull())
plt.title("Missing Values Heatmap")
plt.show()

# Convert Release_Date to datetime and extract Year
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Year'] = df['Release_Date'].dt.year


# Q. 1) For 'House of Cards', what is the Show Id and Who is the Director?
house_cards = df[df['Title'] == 'House of Cards']
print("\n--- House of Cards Info ---")
print(house_cards[['Show_Id', 'Director']])

# Q. 2) Highest number of TV Shows & Movies released per year? Bar Graph.
plt.figure(figsize=(10, 5))
df['Year'].value_counts().plot(kind='bar')
plt.title("Content Released per Year")
plt.show()

# Q. 3) How many Movies & TV Shows? Bar Graph.
# In your dataset, 'Category' stores Movie/TV Show
sns.countplot(data=df, x='Category')
plt.title("Content Categories")
plt.show()

# Q. 4) Show all Movies released in year 2000.
movies_2000 = df[(df['Category'] == 'Movie') & (df['Year'] == 2000)]
print(f"\nMovies in 2000: {len(movies_2000)}")

# Q. 5) Titles of TV Shows released in India only.
tv_india = df[(df['Category'] == 'TV Show') & (df['Country'] == 'India')]['Title']
print("\n--- India TV Shows (Top 5) ---")
print(tv_india.head())

# Q. 6) Top 10 Directors
print("\n--- Top 10 Directors ---")
print(df['Director'].value_counts().head(10))

# Q. 7) Category is Movie & Type is Comedies OR Country is UK
logic_filter = df[((df['Category'] == 'Movie') & (df['Type'].str.contains('Comedies'))) | (df['Country'] == 'United Kingdom')]

# Q. 8) In how many movies/shows was Tom Cruise cast?
df_cast = df.dropna(subset=['Cast'])
tom_cruise = df_cast[df_cast['Cast'].str.contains('Tom Cruise')]
print(f"\nShows with Tom Cruise: {len(tom_cruise)}")

# Q. 9) Different Ratings defined by Netflix
print("\nRatings:", df['Rating'].unique())

# Q. 9.1) TV-14 Movies in Canada
tv14_canada = df[(df['Category'] == 'Movie') & (df['Rating'] == 'TV-14') & (df['Country'] == 'Canada')]

# Q. 9.2) TV Shows with 'R' rating after year 2018
r_tv_2018 = df[(df['Category'] == 'TV Show') & (df['Rating'] == 'R') & (df['Year'] > 2018)]

# Q. 10) Maximum duration of a Movie/Show
df[['Minutes', 'Unit']] = df['Duration'].str.split(' ', expand=True)
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')
print(f"\nMax Duration: {df['Minutes'].max()} Minutes")

# Q. 11) Country with Highest No. of TV Shows
highest_tv = df[df['Category'] == 'TV Show']['Country'].value_counts().head(1)
print("\n--- Country with Most TV Shows ---")
print(highest_tv)

# Q. 12) Sort the dataset by Year
df.sort_values(by='Year', ascending=False, inplace=True)

# Q. 13) Category 'Movie' & Type 'Dramas' OR Category 'TV Show' & Type 'Kids TV'
complex_filter = df[((df['Category'] == 'Movie') & (df['Type'].str.contains('Dramas'))) | 
                   ((df['Category'] == 'TV Show') & (df['Type'].str.contains("Kids' TV")))]
```
