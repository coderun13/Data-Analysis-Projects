```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_csv('london_housing_data.csv')

# Q. 1) Convert 'Date' column to Date-Time format
df['date'] = pd.to_datetime(df['date']) 

# Q. 2) Add a new column 'year'
df['year'] = df['date'].dt.year

# (2.B) Add 'month' as 2nd column
df.insert(1, 'month', df['date'].dt.month)

# Q. 4) Show records where 'No. of Crimes' is 0
zero_crimes_df = df[df.no_of_crimes == 0]
print(f"\nNumber of records with 0 crimes: {len(zero_crimes_df)}")

# Q. 5) Max & Min 'average_price' per year in England
england_df = df[df.area == 'england']
price_stats = england_df.groupby('year').average_price.agg(['max', 'min'])
print("\n--- England Price Stats per Year ---")
print(price_stats)

# Q. 6) Max & Min No. of Crimes recorded per area
crime_stats = df.groupby('area').no_of_crimes.agg(['max', 'min'])
print("\n--- Crime Stats per Area ---")
print(crime_stats)

# Q. 7) Total count of records per area where price < 100,000
low_price_counts = df[df.average_price < 100000].area.value_counts()
print("\n--- Low Price Counts (<100k) ---")
print(low_price_counts)

# Q. 3) Remove the columns 'year' and 'month' (Permanent drop)
df.drop(['year', 'month'], axis=1, inplace=True)
print("\nColumns after dropping year/month:", df.columns)
```
