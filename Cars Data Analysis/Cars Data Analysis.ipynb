```python
import pandas as pd

car = pd.read_csv('Cars Data.csv')

print("Dataset Overview:")
print(car.head())
print("\nMissing values before cleaning:")
print(car.isnull().sum())

# Q. 1) DATA CLEANING: Handle Null Values
car.fillna(car.mean(numeric_only=True), inplace=True)


# Q. 2) VALUE COUNTS: Different 'Make' and their occurrences
make_counts = car['Make'].value_counts()
print("\nCounts for each Car Make:")
print(make_counts)

# Q. 3) FILTERING: Origin is Asia or Europe
asia_europe_cars = car[car['Origin'].isin(['Asia', 'Europe'])]


# Q. 4) REMOVING RECORDS: Weight above 4000
# remove by filtering for records that are LESS than or EQUAL to 4000
car = car[~(car['Weight'] > 4000)]
# Alternative: car = car[car['Weight'] <= 4000]


# Q. 5) APPLYING FUNCTIONS: Increase MPG_City by 3
car['MPG_City'] = car['MPG_City'].apply(lambda x: x + 3)

print("\n--- Final Checks ---")
print(f"New Shape after removing heavy cars: {car.shape}")
print(f"Updated MPG_City (first 5):\n{car['MPG_City'].head()}")
```
