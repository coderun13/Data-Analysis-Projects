```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. LOAD DATA
df = pd.read_csv('Spotify-YouTube.csv')

# --- DATA CLEANING ---
# Standardize column names to avoid KeyErrors
df.columns = df.columns.str.strip()

# Ensure numeric columns are actually numbers
numeric_cols = ['Views', 'Likes', 'Comments', 'Stream', 'Danceability']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Q.1) Top 10 Artists - Highest Views on YouTube
top_10_artists_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10)
print("--- Top 10 Artists by YouTube Views ---")
print(top_10_artists_views)

# Q.2) Top 10 Tracks - Highest Streams on Spotify
top_10_tracks_streams = df.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)
print("\n--- Top 10 Tracks by Spotify Streams ---")
print(top_10_tracks_streams)

# Q.3) Most common Album Types on Spotify
album_types = df['Album_type'].value_counts()
print("\n--- Album Type Counts ---")
print(album_types)

# Q.4) Average Views, Likes, and Comments compared between Album Types
album_comparison = df.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean()
print("\n--- Average Engagement per Album Type ---")
print(album_comparison)

# Q.5) Top 5 YouTube Channels based on Views
top_5_channels = df.groupby('Channel')['Views'].sum().sort_values(ascending=False).head(5)
print("\n--- Top 5 YouTube Channels ---")
print(top_5_channels)

# Q.6) The Top Most Track based on Views
top_track = df[df['Views'] == df['Views'].max()][['Track', 'Artist', 'Views']]
print("\n--- Most Viewed Track ---")
print(top_track)

# Q.7) Top 7 Tracks with highest Like-to-View ratio
# Note: We add a small constant to denominator to avoid division by zero
df['Like_to_View_Ratio'] = df['Likes'] / (df['Views'] + 1)
top_7_ratio = df.sort_values(by='Like_to_View_Ratio', ascending=False).head(7)[['Track', 'Like_to_View_Ratio']]
print("\n--- Top 7 Tracks by Like-to-View Ratio ---")
print(top_7_ratio)

# Q.8) Top Albums having the Tracks with Maximum Danceability
top_dance_albums = df.sort_values(by='Danceability', ascending=False).head(10)[['Album', 'Track', 'Danceability']]
print("\n--- Top 10 Most Danceable Tracks/Albums ---")
print(top_dance_albums)

# Q.9) Correlation between Views, Likes, Comments, and Stream
correlation_matrix = df[['Views', 'Likes', 'Comments', 'Stream']].corr()
print("\n--- Correlation Matrix ---")
print(correlation_matrix)

# Visualization of Correlation
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap: YouTube vs Spotify Metrics")
plt.show()
```
