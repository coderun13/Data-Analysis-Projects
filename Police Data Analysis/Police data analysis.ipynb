```python
import pandas as pd
df = pd.read_csv('Police Data.csv')

# Q. 1) DATA CLEANING: Remove column with ONLY missing values
df.dropna(axis=1, how='all', inplace=True)

# Q. 2) FILTERING + VALUE COUNTS: Speeding by Gender
speeding_gender = df[df['violation'] == 'Speeding'].driver_gender.value_counts()
print("Speeding Stops by Gender:\n", speeding_gender)

# Q. 3) GROUPBY: Does gender affect searches?
# 'search_conducted' is usually a Boolean (True/False)
search_by_gender = df.groupby('driver_gender').search_conducted.sum()
print("\nTotal Searches Conducted by Gender:\n", search_by_gender)

# Q. 4) MAPPING + CASTING: Mean stop_duration
# 'stop_duration' often contains strings like '0-15 min', '16-30 min'. 
# We must map these to integers (mean values) to calculate the overall mean.

# First, check the unique values to see what to map
# print(df['stop_duration'].unique())

df['stop_duration'] = df['stop_duration'].map({'0-15 min': 7.5, '16-30 min': 24, '30+ min': 45})

# Calculate the mean
mean_stop_time = df['stop_duration'].mean()
print(f"\nThe Mean Stop Duration is: {mean_stop_time:.2f} minutes")

# Q. 5) GROUPBY + DESCRIBE: Age distribution by Violation
# This gives a full statistical summary (mean, std, min, max, quartiles)
age_stats = df.groupby('violation').driver_age.describe()
print("\nAge Distribution Summary by Violation:")
print(age_stats)
```
