```python
import pandas as pd

# Load data
df = pd.read_csv('Udemy Courses data.csv')

df['price'] = df['price'].replace('Free', 0)
df['price'] = pd.to_numeric(df['price'], errors='coerce').fillna(0)

# Q. 1) Different subjects
print("Subjects:", df['subject'].unique())

# Q. 2) Max courses subject
print("\nMax Courses:", df['subject'].value_counts().head(1))

# Q. 3 & 4) Free vs Paid
print(f"Free: {len(df[df['is_paid'] == False])}, Paid: {len(df[df['is_paid'] == True])}")

# Q. 5 & 6) Top/Least Selling
top_selling = df.sort_values('num_subscribers', ascending=False)

# Q. 7) Graphic Design price below 100
graphic_design_cheap = df[(df['subject'] == 'Graphic Design') & (df['price'] < 100)]
print(f"\nCheap Graphic Design Courses: {len(graphic_design_cheap)}")

# Q. 8) Python Courses
python_courses = df[df['course_title'].str.contains('Python', case=False)]

# Q. 9) Year 2015
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
courses_2015 = df[df['published_timestamp'].dt.year == 2015]
print(f"2015 Courses: {len(courses_2015)}")

# Q. 10) Max Subscribers per Level
print("\nMax Subs per Level:\n", df.groupby('level')['num_subscribers'].max())
```
