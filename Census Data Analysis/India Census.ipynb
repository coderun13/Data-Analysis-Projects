```python
import pandas as pd

# 1. LOAD DATA
df = pd.read_csv('India Census 2011.csv')

# Q. 1 & 2) Show first 5 records, hide index, and set a caption
styled_df = df.head().style.hide().set_caption('Census 2011 Data Description')
display(styled_df)


# Q. 3) Records for specific districts
districts_list = ['New Delhi', 'Lucknow', 'Jaipur']
specific_districts = df[df['District_name'].isin(districts_list)]
print("\n--- Specific Districts ---")
print(specific_districts)


# Q. 4) State-wise totals
# A. Total Population
state_pop = df.groupby('State_name')['Population'].sum()

# B. Religion-wise Population
# Identifying religion columns (adjust names based on your specific CSV headers)
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
state_religions = df.groupby('State_name')[religion_cols].sum()

print("\n--- State-wise Population ---")
print(state_pop)
print("\n--- State-wise Religion Data ---")
print(state_religions)

# Q. 5) Male Workers in Maharashtra
# Filtering State first, then selecting the column to sum
mah_male_workers = df[df['State_name'] == 'MAHARASHTRA']['Male_Workers'].sum()
print(f"\nTotal Male Workers in Maharashtra: {mah_male_workers}")

# Q. 6) Set 'District_name' as index
df.set_index('District_name', inplace=True)
print("\n--- New Index Set ---")
print(df.head(2))

# Q. 7a) Add Suffix to column names
df = df.add_suffix('_value')

# Q. 7b) Add Prefix to column names
df = df.add_prefix('census_')

print("\n--- Final Column Names (with Prefix & Suffix) ---")
print(df.columns)
```
